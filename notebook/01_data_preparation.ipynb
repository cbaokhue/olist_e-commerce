```python
import pandas as pd
import os

customers = pd.read_csv('olist_customers_dataset.csv')
geolocation = pd.read_csv('olist_geolocation_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
order_payments = pd.read_csv('olist_order_payments_dataset.csv')
order_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')
```

## Checking for missing values and duplicates

```python
def inspect(df, name):
    print(f'--- {name} ---')
    print(f'Shape: {df.shape}')
    print(f'Duplicates: {df.duplicated().sum()}')
    print(f'Missing values:\n{df.isnull().sum()}')
    print()

# Inspecting each tables
inspect(customers, 'customers')
inspect(geolocation, 'geolocation')
inspect(order_items, 'order_items')
inspect(order_payments, 'order_payments')
inspect(order_reviews, 'order_reviews')
inspect(orders, 'orders')
inspect(products, 'products')
inspect(sellers, 'sellers')
inspect(translation, 'translation')
```

```python
# Remove duplicated rows
geolocation = geolocation.drop_duplicates()

# Deduplicate by zip code prefix
geolocation = geolocation.drop_duplicates(subset='geolocation_zip_code_prefix', keep='first')

print(f'Cleaned shape: {geolocation.shape}')
```

## Fixing mispelled columns

```python
# Rename mispelled columns
products = products.rename(columns={
    'product_name_lenght': 'product_name_length',
    'product_description_lenght': 'product_description_length'
})

print(products.columns.tolist())
```

## Checking for unusual data types

```python
def data_types(df, name):
    print(f'--- {name} ---')
    print(f'{df.dtypes}')
    print()

data_types(customers, 'customers')
data_types(geolocation, 'geolocation')
data_types(order_items, 'order_items')
data_types(order_payments, 'order_payments')
data_types(order_reviews, 'order_reviews')
data_types(orders, 'orders')
data_types(products, 'products')
data_types(sellers, 'sellers')
data_types(translation, 'translation')
```

```python
# order_items
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])

# order_reviews
order_reviews[['review_creation_date', 'review_answer_timestamp']] = order_reviews[['review_creation_date', 'review_answer_timestamp']].apply(pd.to_datetime)

# orders
time_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
orders[time_cols] = orders[time_cols].apply(pd.to_datetime)

# products
int_cols = ['product_name_length', 'product_description_length', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
products[int_cols] = products[int_cols].astype('Int64')
```

```python
data_types(customers, 'customers')
data_types(geolocation, 'geolocation')
data_types(order_items, 'order_items')
data_types(order_payments, 'order_payments')
data_types(order_reviews, 'order_reviews')
data_types(orders, 'orders')
data_types(products, 'products')
data_types(sellers, 'sellers')
data_types(translation, 'translation')
```
